```python
import pandas as pd
import json
import requests
import io
aarTilSSB = "2020"
kommuneID = "0301"
kommuneNavn = "drammen"
semesterTid = "3"
```

```python
# tabell 662: https://dbh.nsd.uib.no/dokumentasjon/tabell.action?tabellId=662
print("Henter campusinformasjon fra NSD... \n")
data = {"tabell_id":662,"api_versjon":1,"statuslinje":"N","begrensning":"10000","kodetekst":"J","desimal_separator":".",
"variabler":["*"],
"sortBy":["Årstall", "Semester"],
"filter":[
   {
      "variabel": "Årstall",
      "selection": {
         "filter": "item",
         "values": [
            aarTilSSB
         ],
         "exclude": [
            ""
         ]
      }
   },
   {
      "variabel": "Semester",
      "selection": {
         "filter": "item",
         "values": [
            semesterTid
         ],
         "exclude": [
            ""
         ]
      }
   },
   {
      "variabel": "Kommunenummer campus",
      "selection": {
         "filter": "item",
         "values": [
            kommuneID
         ],
         "exclude": [
            ""
         ]
      }
   }
]} 
response = requests.post("https://api.nsd.no/dbhapitjener/Tabeller/hentJSONTabellData", json=data)
if response.status_code == 200:
    if semesterTid == "3":
        print("OK. Responskode 200 - tabell hentet for høstsemesteret", aarTilSSB +".")
    elif semesterTid == "1":
        print("OK. Responskode 200 - tabell hentet for vårsemesteret", aarTilSSB +".")
    else:
        print("... Altså, nå har du klart å søke på et semester som ikke er vår- eller høstsemester. Likevel fikk du respons 200 OK. ???")
elif response.status_code == 201:
    print("Responskode 201 - Created? .. Ja, det kan jo hende.\n")
elif response.status_code == 204:
    print("Ingen resultater for denne kommunen.\n")
else:
    print("Responskode", response.status_code +". Noe er feil, eller så kan det hende serveren er nede.\n")
# Assign responsen til en variabel, deretter anvend io for å konvertere til pandas dataframe.

nsdtabeller = json.loads(response.text)
```

```python
z = 0
print("Viser campus..\n")
for i in nsdtabeller:
    z = z + 1
    print("Innførsel", z)
    print("Navn:", i['Institusjonsnavn'])
    print("Campus:", i["Stednavn campus"])
    print("Kommunenummer:", i["Kommunenummer campus"])
    print('\n')
if z != 0:
        print("Totalt", z, "campus.")
else:
    print("Ingen campusinformasjon å vise for", kommuneNavn.capitalize() +".")
```
