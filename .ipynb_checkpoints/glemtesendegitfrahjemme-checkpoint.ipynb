```python
import pandas as pd
import json
import requests
import io
aarTilSSB = "2020"
kommuneID = "0301"
kommuneNavn = "drammen"
```

```python
# tabell 662: https://dbh.nsd.uib.no/dokumentasjon/tabell.action?tabellId=662
data = {"tabell_id":662,"api_versjon":1,"statuslinje":"N","begrensning":"10000","kodetekst":"J","desimal_separator":".",
"variabler":["*"],
"sortBy":["Årstall", "Semester"],
"filter":[
   {
      "variabel": "Årstall",
      "selection": {
         "filter": "item",
         "values": [
            aarTilSSB
         ],
         "exclude": [
            ""
         ]
      }
   },
   {
      "variabel": "Semester",
      "selection": {
         "filter": "all",
         "values": [
            "*"
         ],
         "exclude": [
            ""
         ]
      }
   },
   {
      "variabel": "Institusjonskode",
      "selection": {
         "filter": "all",
         "values": [
            "*"
         ],
         "exclude": [
            ""
         ]
      }
   }
]} 
response = requests.post("https://api.nsd.no/dbhapitjener/Tabeller/hentJSONTabellData", json=data)
if response.status_code == 200:
    print("OK. Responskode 200 - tabell hentet.")
elif response.status_code == 201:
    print("Responskode 201 - Created? .. Ja, det kan jo hende.")
else:
    print("Responskode", response.status_code +". Noe er feil, ellers er serveren nede.")
# Assign responsen til en variabel, deretter anvend io for å konvertere til pandas dataframe.

nsdtabeller = json.loads(response.text)
```

```python
nsdtabeller
```

```python
print("Henter campusinformasjon fra NSD... \n")
z = 0
for i in nsdtabeller:
    if i['Kommunenummer campus'] == kommuneID:
        z = z + 1
        print("Navn:", i['Institusjonsnavn'])
        print("Campus:", i["Stednavn campus"])
        print("Kommunenummer:", i["Kommunenummer campus"])
        print('\n')
if z != 0:
        print("Campusinformasjon for", z, "campus innhentet.")
else:
    print("Ingen campusinformasjon å hente for", kommuneNavn.capitalize() +".")
```
